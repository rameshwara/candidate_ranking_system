```python
import pandas as pd
import json
from nltk.tokenize import RegexpTokenizer
```

```python
json_file = "/content/drive/MyDrive/Hackthon/IndiaRuns/India_runs_data_and_ai_challenge/candidates.jsonl"

profiles = []
career_history = []
education = []
skills = []
certifications = []
languages = []
redrob_signals = []
industries = []
with open(json_file, "r", encoding="utf-8") as f:
    for line in f:
        candidate = json.loads(line)

        candidate_id = candidate["candidate_id"]

        # -------------------------------
        # Candidate Profile
        # -------------------------------
        profile = candidate.get("profile", {})

        profiles.append({
            "candidate_id": candidate_id,
            "name": profile.get("anonymized_name"),
            "headline": profile.get("headline"),
            "summary": profile.get("summary"),
            "location": profile.get("location"),
            "country": profile.get("country"),
            "years_of_exp": profile.get("years_of_experience"),
            "current_title": profile.get("current_title"),
            "current_company": profile.get("current_company"),
            "current_company_size": profile.get("current_company_size"),
            "current_industry": profile.get("current_industry")
        })

                # Industry from current profile
        industries.append({
            "candidate_id": candidate_id,
            "Type_of_org": profile.get("current_industry"),
            "company_size": profile.get("current_company_size"),

        })

        # -------------------------------
        # Career History
        # -------------------------------
        for exp in candidate.get("career_history", []):
            career_history.append({
                "candidate_id": candidate_id,
                "company": exp.get("company"),
                "title": exp.get("title"),
                "start_date": exp.get("start_date"),
                "end_date": exp.get("end_date"),
                "duration_months": exp.get("duration_months"),
                "is_current": exp.get("is_current"),
                "industry": exp.get("industry"),
                "company_size": exp.get("company_size"),
                "description": exp.get("description")
            })


            # Industry from career history
            industries.append({
                "candidate_id": candidate_id,
                "Type_of_org": exp.get("industry"),
                "company_size": exp.get("company_size"),
            })

        # -------------------------------
        # Education
        # -------------------------------
        for edu in candidate.get("education", []):
            education.append({
                "candidate_id": candidate_id,
                "institution": edu.get("institution"),
                "degree": edu.get("degree"),
                "field_of_study": edu.get("field_of_study"),
                "start_year": edu.get("start_year"),
                "end_year": edu.get("end_year"),
                "grade": edu.get("grade"),
                "tier": edu.get("tier")
            })

        # -------------------------------
        # Skills
        # -------------------------------
        for skill in candidate.get("skills", []):
            skills.append({
                "candidate_id": candidate_id,
                "name": skill.get("name"),
                "proficiency": skill.get("proficiency"),
                "endorsements": skill.get("endorsements"),
                "duration_months": skill.get("duration_months")
            })

        # -------------------------------
        # Certifications
        # -------------------------------
        for cert in candidate.get("certifications", []):
            certifications.append({
                "candidate_id": candidate_id,
                "name": cert.get("name"),
                "issuer": cert.get("issuer"),
                "year": cert.get("year")
            })

        # -------------------------------
        # Languages
        # -------------------------------
        for lang in candidate.get("languages", []):
            languages.append({
                "candidate_id": candidate_id,
                "language": lang.get("language"),
                "proficiency": lang.get("proficiency")
            })

        # -------------------------------
        # Redrob Signals
        # -------------------------------
        signal = candidate.get("redrob_signals", {})

        redrob_signals.append({
            "candidate_id": candidate_id,
            "profile_completeness_score": signal.get("profile_completeness_score"),
            "signup_date": signal.get("signup_date"),
            "last_active_date": signal.get("last_active_date"),
            "open_to_work_flag": signal.get("open_to_work_flag"),
            "profile_views_received_30d": signal.get("profile_views_received_30d"),
            "applications_submitted_30d": signal.get("applications_submitted_30d"),
            "recruiter_response_rate": signal.get("recruiter_response_rate"),
            "avg_response_time_hours": signal.get("avg_response_time_hours"),
            "skill_assessment_scores": json.dumps(
                signal.get("skill_assessment_scores", {})
            ),
            "connection_count": signal.get("connection_count"),
            "endorsements_received": signal.get("endorsements_received"),
            "notice_period_days": signal.get("notice_period_days"),
            "expected_min_salary_lpa":
                signal.get("expected_salary_range_inr_lpa", {}).get("min"),
            "expected_max_salary_lpa":
                signal.get("expected_salary_range_inr_lpa", {}).get("max"),
            "preferred_work_mode": signal.get("preferred_work_mode"),
            "willing_to_relocate": signal.get("willing_to_relocate"),
            "github_activity_score": signal.get("github_activity_score"),
            "search_appearance_30d": signal.get("search_appearance_30d"),
            "saved_by_recruiters_30d": signal.get("saved_by_recruiters_30d"),
            "interview_completion_rate": signal.get("interview_completion_rate"),
            "offer_acceptance_rate": signal.get("offer_acceptance_rate"),
            "verified_email": signal.get("verified_email"),
            "verified_phone": signal.get("verified_phone"),
            "linkedin_connected": signal.get("linkedin_connected")
        })






# ==========================================
# Create CSV files
# ==========================================
datasetPath = "/content/drive/MyDrive/Hackthon/IndiaRuns/India_runs_data_and_ai_challenge/dataset/"
pd.DataFrame(profiles).to_csv(
    datasetPath+"candidate_profile.csv", index=False)

pd.DataFrame(career_history).to_csv(
    datasetPath+"candidate_career_history.csv", index=False)

pd.DataFrame(education).to_csv(
    datasetPath+"candidate_education.csv", index=False)

pd.DataFrame(skills).to_csv(
    datasetPath+"candidate_skills.csv", index=False)

pd.DataFrame(certifications).to_csv(
    datasetPath+"candidate_certifications.csv", index=False)

pd.DataFrame(languages).to_csv(
    datasetPath+"candidate_languages.csv", index=False)

pd.DataFrame(redrob_signals).to_csv(
    datasetPath+"candidate_redrob_signals.csv", index=False)

pd.DataFrame(industries).to_csv(
    datasetPath+"candidate_org.csv", index=False)
print("All CSV files generated successfully.")
```

```python

# ---------------------------------
# Institution Rankings
# ---------------------------------

ranking_map = {
    "Tier-3 Engineering College": (5000, 10000),
    "Local Engineering College": (4500, 10000),
    "Generic State University": (300, 10000),
    "Regional Technical Institute": (250, 10000),
    "Amity University": (40, 1200),
    "Lovely Professional University": (35, 900),
    "SRM Chennai": (18, 1000),
    "Bharati Vidyapeeth": (90, 10000),
    "Symbiosis International": (45, 800),
    "VIT Chennai": (25, 850),
    "Chandigarh University": (30, 800),
    "KIIT University": (40, 1100),
    "Christ University": (75, 1300),
    "COEP Pune": (65, 10000),
    "RV College of Engineering": (85, 10000),
    "SRM University": (15, 950),
    "Anna University": (14, 400),
    "VJTI Mumbai": (70, 10000),
    "Delhi College of Engineering": (27, 800),
    "Manipal Institute of Technology": (23, 900),
    "Thapar University": (29, 850),
    "PES University": (80, 10000),
    "VIT Vellore": (19, 800),
    "Jadavpur University": (17, 750),
    "IIT Roorkee": (6, 350),
    "IIT Delhi": (2, 125),
    "IIT Kharagpur": (5, 225),
    "IIT Bombay": (3, 120),
    "Carnegie Mellon University": (1, 50),
    "Stanford University": (1, 5),
    "IIT Madras": (1, 200),
    "IIIT Hyderabad": (40, 700),
    "IISc Bangalore": (1, 220),
    "NIT Warangal": (28, 1000),
    "NIT Surathkal": (17, 900),
    "BITS Pilani": (11, 800),
    "Massachusetts Institute of Technology": (1, 1),
    "Georgia Tech": (5, 100),
    "UC Berkeley": (3, 10),
    "IIT Guwahati": (8, 400),
    "NIT Trichy": (9, 850),
    "IIT Kanpur": (4, 275),
    "IIIT Bangalore": (60, 10000),
    "IIT Hyderabad": (7, 600)
}

# Default for unknown institutions
DEFAULT_NATIONAL = 5000
DEFAULT_GLOBAL = 10000

edu_df["national_ranking"] = edu_df["institution"].map(
    lambda x: ranking_map.get(x, (DEFAULT_NATIONAL, DEFAULT_GLOBAL))[0]
)

edu_df["global_ranking"] = edu_df["institution"].map(
    lambda x: ranking_map.get(x, (DEFAULT_NATIONAL, DEFAULT_GLOBAL))[1]
)

# ---------------------------------
# Industry Connections Score
# ---------------------------------

def industry_connections(rank):
    if rank <= 10:
        return 10
    elif rank <= 25:
        return 9
    elif rank <= 50:
        return 8
    elif rank <= 100:
        return 7
    elif rank <= 500:
        return 5
    else:
        return 2

edu_df["industry_connections"] = edu_df["national_ranking"].apply(
    industry_connections
)

# ---------------------------------
# Quality of Course Score
# ---------------------------------

def quality_of_course(field):

    field = str(field).lower()

    premium = [
        "computer science",
        "artificial intelligence",
        "machine learning",
        "data science",
        "software engineering",
        "electronics",
        "electrical",
        "information technology"
    ]

    good = [
        "mechanical",
        "civil",
        "chemical",
        "industrial",
        "mathematics",
        "physics"
    ]

    if any(x in field for x in premium):
        return 10

    if any(x in field for x in good):
        return 8

    return 6

edu_df["quality_of_course"] = edu_df["field_of_study"].apply(
    quality_of_course
)

# ---------------------------------
# Final Output
# ---------------------------------

output_df = edu_df[
    [
        "candidate_id",
        "institution",
        "national_ranking",
        "global_ranking",
        "tier",
        "industry_connections",
        "quality_of_course"
    ]
]

output_df.to_csv(
    datasetPath+"candidate_education_enriched.csv",
    index=False
)

print(output_df.head())
```

```python
df_cert=pd.read_csv(datasetPath+"candidate_certifications.csv")
```

```python


CURRENT_YEAR = 2026

# =====================================================
# Issuer Master Data
# =====================================================

issuer_info = {
    "AWS": {
        "base_location": "United States",
        "established_year": 2006,
        "recognition": 97
    },
    "ASQ": {
        "base_location": "United States",
        "established_year": 1946,
        "recognition": 85
    },
    "Scrum Alliance": {
        "base_location": "United States",
        "established_year": 2001,
        "recognition": 88
    },
    "Coursera/DeepLearning.AI": {
        "base_location": "United States",
        "established_year": 2017,
        "recognition": 90
    },
    "Google Cloud": {
        "base_location": "United States",
        "established_year": 2008,
        "recognition": 98
    },
    "DeepLearning.AI": {
        "base_location": "United States",
        "established_year": 2017,
        "recognition": 92
    }
}

# =====================================================
# Feature Functions
# =====================================================

def get_base_location(issuer):
    issuer = str(issuer).strip()

    if issuer in issuer_info:
        return issuer_info[issuer]["base_location"]

    return "Unknown"


def get_issue_location(issuer):
    return get_base_location(issuer)


def get_establishment_age(issuer):
    issuer = str(issuer).strip()

    if issuer in issuer_info:
        return CURRENT_YEAR - issuer_info[issuer]["established_year"]

    return None


def get_recognition(issuer):
    issuer = str(issuer).strip()

    if issuer in issuer_info:
        return issuer_info[issuer]["recognition"]

    return 50


# =====================================================
# Create Feature Dataset
# =====================================================

output_df = pd.DataFrame()

output_df["candidate_id"] = df_cert["candidate_id"]
output_df["issuer"] = df_cert["issuer"]

output_df["issuer_base_location"] = (
    df_cert["issuer"].apply(get_base_location)
)

output_df["issue_location"] = (
    df_cert["issuer"].apply(get_issue_location)
)

output_df["issuer_establishment_age"] = (
    df_cert["issuer"].apply(get_establishment_age)
)

output_df["issuer_recognition_in_community"] = (
    df_cert["issuer"].apply(get_recognition)
)

# =====================================================
# Save CSV
# =====================================================

output_df.to_csv(
    datasetPath+"certification_issuer_features.csv",
    index=False
)

print(output_df.head())

print("\nShape:", output_df.shape)

print("\nIssuer Distribution:")
print(output_df["issuer"].value_counts())
```
